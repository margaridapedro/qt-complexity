```python
import requests
import re
from bs4 import BeautifulSoup

page = requests.get("https://web.archive.org/web/20181208091423/https://math.nist.gov/quantum/zoo/")
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
all_b = soup.body.find_all("b")
```

```python
problems = {}

#<script type="math/tex" id="MathJax-Element-3"> O(2^{n/4+o(1)}) </script>

for b in all_b:
    if("Algorithm:" == b.getText()):
        it = b
        name = str(it.next_element.next_element).strip()
        print("Algorithm:", name)
        it = it.find_next_sibling("b")
        speedup_qltt = str(it.next_element.next_element).strip()
        print("Speed up:", speedup_qltt)
        it = it.find_next_sibling("b").next_element # move to description
        problems[name] = {"speedup": speedup_qltt, "formulas": []}
        description = ""
        while it.next_element.name not in ["b", "br"]:
            it = it.next_element
            description += str(it).strip()
        formulas = re.findall('(O\(.+?\))', description)
        formulas2 = re.findall('(2\^\(.+?\))', description)
        print(formulas)
        print(formulas2)
        problems[name]["formulas"] = formulas        
```
